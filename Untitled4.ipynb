```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.m.wikipedia.org/wiki/Economy_of_the_United_Kingdom'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('table', class_ = 'wikitable')
```

```python
table = soup.find_all('table')[1]
```

```python
sector_titles = table.find_all('th')
print(sector_titles)
```

```python
sector_table_titles = [title.text.strip() for title in sector_titles]
print(sector_table_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns= sector_table_titles)
df
```

```python
sector_roles = table.find_all('tr')
```

```python
for row in sector_roles[1:]:
    row_data = row.find_all('td')
    ind_row_data = [role.text.strip() for role in row_data]

    length = len(df)
    df.loc[length] = ind_row_data
```

```python
df
```

```python
table3 = soup.find_all('table')[3]
Exchange_titles = table3.find_all('th')
print(Exchange_titles)
```

```python
Exchange_table_titles = [title.text.strip() for title in Exchange_titles]
print(Exchange_table_titles)
```

```python
ef = pd.DataFrame(columns= Exchange_table_titles)
ef
```

```python
exchange_roles = table3.find_all('tr')
```

```python
# for row in exchange_roles[2:]:
#   row_data2 = row.find_all('td')
#   ind_row_data2 = [role.text.strip() for role in row_data2]

#   if ind_row_data2:  # This will check if the list is non-empty
#         length = len(ef)
#         ef.loc[length] = ind_row_data2

# Assuming ef is already defined and initialized as a DataFrame with appropriate columns
column_count = len(ef.columns)

for row in exchange_roles[2:]:
    row_data2 = row.find_all('td')
    ind_row_data2 = [role.text.strip() for role in row_data2]

    if ind_row_data2:  # Ensure it's not an empty list
        if len(ind_row_data2) == column_count:
            ef.loc[len(ef)] = ind_row_data2
        elif len(ind_row_data2) < column_count:
            # Pad the row with None or appropriate values to match the number of columns
            ind_row_data2.extend([None] * (column_count - len(ind_row_data2)))
            ef.loc[len(ef)] = ind_row_data2
        else:
            # Handle the case where there are more elements than columns (optional)
            ind_row_data2 = ind_row_data2[:column_count]
            ef.loc[len(ef)] = ind_row_data2
```
